```python
import pandas as pd
```

```python
# Frame A
cols=["使用者序號","電影序號","比率","時間戳記"]
rating = pd.read_csv('ml-100k/u.data',sep="\t",
                   names=cols)
rating
```

```python
# Frame B
col2=["電影序號","標題","釋出日期",'電影釋出日','宣傳網址']
movies = pd.read_csv('ml-100k/u.item',sep="|",
                   names=col2,
                   usecols=range(5),
                   encoding="ISO-8859-1" )
movies
```

```python
# Frame C = Frame A + Frame B
movies_rating = pd.merge(movies,rating)
movies_rating.head()
```

```python
# Frame D
user_cols=['使用者序號','年齡','性別','職業','郵局代碼']
users = pd.read_csv('ml-100k/u.user',sep="|",names=user_cols)
users.head()
```

```python
# Frame E = Frame C + Frame D
data = pd.merge(movies_rating,users)
data.head()
```

```python
# 以降續方式查詢前20筆,最多人看的電影名稱
most_rated = data.groupby('標題').size().sort_values(ascending=False)[0:20]
most_rated
```

```python
# 以降續方式查詢前20筆,最多人看的電影名稱
most_rated = data.標題.value_counts()[:20]
most_rated
```

```python
import numpy as np
#聚合函数，只需将其传入aggreate或agg方法
#np.size: ndarray 的屬性, ndarray 的元素的數量.
#np.mean: 計算數組的均值,單獨計算每行或每列的均值.
movies_status = data.groupby('標題').agg({'比率':[np.size,np.mean]})
movies_status.head()
```

```python
movies_status.sort_values([('比率','mean')],ascending=False).head()
```

```python
import matplotlib.pyplot as plt
users.年齡.hist(bins = 30)
plt.title("使用者年齡分布")
plt.xlabel('年齡')
plt.ylabel('使用者人數')
plt.show()
```

```python
most_50 = data.groupby("電影序號").size().sort_values(ascending = False)[:50]
most_50
```

```python
pivoted = data.pivot_table(index=['電影序號','標題'], 
                           columns=["性別"],
                           values='比率',
                           fill_value=0)
pivoted.head()
```

```python
pivoted['diff'] = pivoted.M - pivoted.F
pivoted
```

```python
pivoted.reset_index('電影序號',inplace=True)
pivoted
# 異議 = pivoted[pivoted.isin(most_50.index)]
# 異議.head()
```

```python
pivoted.電影序號
```

```python
#計算出一個布林陣列表示每一個值是否包含在所傳遞的集合裡
異議 = pivoted[pivoted.電影序號.isin(most_50.index)]['diff']
異議.head()
```

```python
plt.title("男生 Vs. 女生 平均比率\n 之間的差異 \n若數值大於0 代表該電影較受男生喜愛")
plt.xlabel("平均比率差異")
plt.ylabel("標題")
異議.plot(kind='barh')
plt.savefig("平均比率差異.png")
plt.show()
```
