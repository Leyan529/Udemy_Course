```python
import re
import urllib.request
```

```python
city = input("Please Enter your city : ")
url = "http://www.weather-forecast.com/locations/"+city+"/forecasts/latest"
data = urllib.request.urlopen(url).read().decode("utf-8")
```

```python
print (data)
```

```python
str1 = 'span class="phrase">'
m1 = re.search(str1,data)
st = m1.start()
so = st + 300
newStr = data[st:so]

str2 =  '</span>'
m2 = re.search(str2,newStr)
final = newStr[len(str1):m2.end()-len(str2)]
print(final)
```
