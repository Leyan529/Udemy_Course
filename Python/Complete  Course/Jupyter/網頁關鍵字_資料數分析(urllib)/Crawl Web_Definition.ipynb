```python
import re
import urllib.request
```

```python
try: 
    url ="http://www.dictionary.com/browse/"
    word = input("Enter your word: ")
    url = url + word
    data = urllib.request.urlopen(url).read().decode("utf-8")
except:
    print("your search word is wrong")
```

```python
print(data)
```

```python
s1='meta name="description" content="'
m = re.search(s1,data)
A = m.start()+len(s1)
newS1 = data[A:]
print(newS1)
```

```python
s2='See more." />'
m = re.search(s2,newS1)
B = m.end()-len(s2)
newS2 = newS1[:B]
print(newS2)
```
