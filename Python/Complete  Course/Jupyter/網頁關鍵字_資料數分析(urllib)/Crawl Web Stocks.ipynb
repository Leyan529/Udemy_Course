```python
#coding:utf-8
import requests,re
from bs4 import BeautifulSoup
```

```python
url = "https://www.google.com.tw/search?q="
```

```python
stock = input("Enter your stock:")
url = url + stock
print(url)
req = requests.get(url)
req.encoding='unicode'
```

```python
# soup = BeautifulSoup(req.text,'lxml')   #創建一個 BeautifulSoup 物件，將網頁讀入
soup = BeautifulSoup(req.content,'lxml') 
content = soup.prettify()
print (content)



```

```python
m = re.search('itemscope=""',soup.prettify())
a = re.search('<title>',soup.prettify())
c = re.search('<div class="sd" id="resultStats">',soup.prettify())  #網頁搜尋比數
```

```python
print(m)
print(a)
print(c) #網頁搜尋比數
```

```python
start_m = m.start()
end_m= start_m+100

start_a = a.start()
end_a= start_a+50

start_c = c.start()  #網頁搜尋比數
end_c= start_c+70
```

```python
new_sm=content[start_m:end_m]
print (new_sm)
```

```python
new_sa=content[start_a:end_a]
print (new_sa)
```

```python
new_sc=content[start_c:end_c]  #網頁搜尋比數
print (new_sc)
```

```python
add_search_m = re.search('lang',new_sm)
start_m = 0
end_m= add_search_m.end()-4
print(new_sm[start_m:end_m])
```

```python
add_search_a = re.search('</title>',new_sa)
start_a = 8
end_a= add_search_a.end()-11
print(new_sa[start_a:end_a])
```

```python
add_search_c = re.search('</div>',new_sc)  #網頁搜尋比數
start_c = 42
end_c= add_search_c.end()-6

print("關鍵字: "+stock +", 網路搜尋比數: "+new_sc[start_c:end_c])
```
