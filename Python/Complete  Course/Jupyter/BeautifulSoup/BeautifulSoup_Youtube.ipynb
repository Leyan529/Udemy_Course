```python
import requests
from bs4 import BeautifulSoup
```

```python
stock = input("請輸入搜尋關鍵字 : ")
url = "https://www.youtube.com/results?search_query=" + stock
data = requests.get(url)
soup = BeautifulSoup(data.text,'html.parser');
```

```python
soup.body.prettify
```

```python
div = soup.find_all('div',{'class' : 'yt-lockup yt-lockup-tile yt-lockup-video vve-check clearfix'})
div
#soup.find('a',{'id' :'video-title'})
```

```python
name_list=[]
href_list = []
for item in div:
    # class="yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink spf-link "
    title_list1 = item.find('a',{'dir' : 'ltr'})
    title_list2 = item.find('a',{'aria-hidden' : 'true'})
    title_list2.find('img').decompose()
    title_list2.find('div',{'class' : 'yt-thumb video-thumb'}).decompose()
    name_list.append(title_list1)
    href_list.append(title_list2)
```

```python
for item in name_list:
    #print(item.get('title'))
    print(item.string)
    print(item.contents)
    print('https://www.youtube.com' + item.get('href'))
```
