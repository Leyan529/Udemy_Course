```python
import urllib.request
from bs4 import BeautifulSoup

# A
url = "http://www.weather-forecast.com/locations/Taipei"
data = urllib.request.urlopen(url).read()
soup = BeautifulSoup(data,"html.parser")
```

```python
# A
print(soup.prettify())
```

```python
# B
import requests
url = "https://mrmad.com.tw/outlast-deluxe"
req = requests.get(url)
soup = BeautifulSoup(req.text,"html.parser")
```

```python
all_data = soup.prettify()
print(all_data)
```

```python
title_tag = soup.title.prettify()
print(title_tag)
```

```python
head_tag = soup.head.prettify()
print(head_tag)
```

```python
body_tag = soup.body.prettify()
print(body_tag)
```

```python
# C tag
print(soup.body.a)
print(soup.body.a.i)
```

```python
# D find all html tags
array = soup.find_all('a')
for i in range(0,len(array)):
    if i<6 :
      array.pop
    elif i>15 and i<30:
        print(array[i])
    
```

```python
# E .[PART].contents
head_tag = soup.head
head_tag.contents
```

```python
body_tag = soup.body
# .contents 和.children 屬性僅包含tag的直接子節點
for i in body_tag.children:
    print(i)
```

```python
for i in body_tag.descendants:
#.descendants 屬性可以對所有tag的子孫節點進行遞歸循環
    print(i)
```
