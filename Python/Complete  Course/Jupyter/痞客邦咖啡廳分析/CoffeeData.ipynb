```python
#coding:utf-8
import requests
from bs4 import BeautifulSoup
url = "http://vilo92.pixnet.net/blog/post/429955862-%E3%80%90%E5%8F%B0%E5%8C%97%E3%80%91%E7%A7%81%E8%97%8F%E5%BF%85%E5%8E%BB%E7%9A%8415%E9%96%93%E7%BE%8E%E5%A5%BD%E5%92%96%E5%95%A1%E5%BB%B3%E2%80%A7%E5%81%B7%E6%B8%A1%E8%B6%85"
req = requests.get(url)
req.encoding='utf-8'
```

```python
soup = BeautifulSoup(req.text,'lxml')   #創建一個 BeautifulSoup 物件，將網頁讀入
print(soup)
```

```python
x = soup.select("#article-content-inner")[0]   #使用select語法 => BeautifulSoup.select("選擇器")
data=[]
table=[]
for index,element in enumerate(x.select("p")):
    if index < 7 or index > 881:
        continue
    else:
        if element.text=="\xa0":
            data.append(table)
            table=[]
            
        else:
          table.append(element.text.strip())
          print(element.text.strip())


```
